# AcadémicaView — Análisis PySpark
Análisis distribuido del dataset de rendimiento académico (n=1220 estudiantes).
Incluye: ETL, estadísticas, regresión lineal (Modelo 1 y Modelo 2) y clustering K-Means.

```python
import requests
from pyspark.sql import SparkSession

def reset_spark():
    try:
        data = requests.get('http://spark-master:8080/json').json()
        for app in data['activeapps']:
            requests.post('http://spark-master:8080/app/kill/', data={'id': app['id']})
    except:
        pass
    return SparkSession.builder \
        .appName('AcademicaView') \
        .master('spark://spark-master:7077') \
        .config('spark.cores.max', '2') \
        .config('spark.jars', '/opt/extra-jars/hadoop-aws-3.3.4.jar,/opt/extra-jars/aws-java-sdk-bundle-1.12.262.jar') \
        .config('spark.driver.extraClassPath', '/opt/extra-jars/*') \
        .config('spark.executor.extraClassPath', '/opt/extra-jars/*') \
        .config('spark.hadoop.fs.s3a.impl', 'org.apache.hadoop.fs.s3a.S3AFileSystem') \
        .config('spark.hadoop.fs.s3a.endpoint', 'http://minio:9000') \
        .config('spark.hadoop.fs.s3a.access.key', 'admin') \
        .config('spark.hadoop.fs.s3a.secret.key', 'password123') \
        .config('spark.hadoop.fs.s3a.path.style.access', 'true') \
        .getOrCreate()

spark = reset_spark()
spark
```

## 1. ETL — Carga y limpieza del dataset

```python
from pyspark.sql import functions as F

df = spark.read.csv(
    's3a://academicaview/datasets/base_imputada_knn.csv',
    header=True,
    inferSchema=True
)

print(f'Registros: {df.count()}')
print(f'Columnas: {len(df.columns)}')
df.printSchema()
```

```python
# Verificar valores nulos
print('Valores nulos por columna:')
df.select([F.count(F.when(F.col(c).isNull(), c)).alias(c) for c in df.columns]).show()

# Estadísticas descriptivas
print('Estadísticas descriptivas:')
df.select('puntaje_final','horas_estudio','promedio_previo','horas_sueno','asistencia','uso_redes').describe().show()
```

## 2. Análisis de Correlaciones

```python
variables = ['puntaje_final','horas_estudio','promedio_previo','horas_sueno','asistencia','uso_redes']

print('Correlaciones con puntaje_final:')
for var in variables[1:]:
    corr = df.stat.corr('puntaje_final', var)
    print(f'  {var}: {corr:.4f}')
```

## 3. Modelo 1 — Regresión Lineal con 4 variables

```python
from pyspark.ml.feature import VectorAssembler, StringIndexer
from pyspark.ml.regression import LinearRegression
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import RegressionEvaluator

indexer = StringIndexer(inputCol='modalidad', outputCol='modalidad_idx')

assembler_1 = VectorAssembler(
    inputCols=['horas_estudio','promedio_previo','horas_sueno','modalidad_idx'],
    outputCol='features',
    handleInvalid='skip'
)

lr = LinearRegression(featuresCol='features', labelCol='puntaje_final')

pipeline_1 = Pipeline(stages=[indexer, assembler_1, lr])

train, test = df.randomSplit([0.8, 0.2], seed=42)
model_1 = pipeline_1.fit(train)
pred_1 = model_1.transform(test)

evaluator_r2   = RegressionEvaluator(labelCol='puntaje_final', metricName='r2')
evaluator_rmse = RegressionEvaluator(labelCol='puntaje_final', metricName='rmse')
evaluator_mae  = RegressionEvaluator(labelCol='puntaje_final', metricName='mae')

r2_1   = evaluator_r2.evaluate(pred_1)
rmse_1 = evaluator_rmse.evaluate(pred_1)
mae_1  = evaluator_mae.evaluate(pred_1)

print('=== MODELO 1 (4 variables) ===')
print(f'R²:   {r2_1:.4f}')
print(f'RMSE: {rmse_1:.4f}')
print(f'MAE:  {mae_1:.4f}')
```

## 4. Modelo 2 — Regresión Lineal (6 variables)

```python
assembler_2 = VectorAssembler(
    inputCols=['horas_estudio','promedio_previo','horas_sueno','modalidad_idx','asistencia','uso_redes'],
    outputCol='features',
    handleInvalid='skip'
)

pipeline_2 = Pipeline(stages=[indexer, assembler_2, lr])
model_2 = pipeline_2.fit(train)
pred_2 = model_2.transform(test)

r2_2   = evaluator_r2.evaluate(pred_2)
rmse_2 = evaluator_rmse.evaluate(pred_2)
mae_2  = evaluator_mae.evaluate(pred_2)

print('=== MODELO 2 (6 variables) — usado en la app ===')
print(f'R²:   {r2_2:.4f}')
print(f'RMSE: {rmse_2:.4f}')
print(f'MAE:  {mae_2:.4f}')

print('\n=== COMPARACIÓN ===')
print(f'{'Métrica':<10} {'Modelo 1':>12} {'Modelo 2':>12}')
print(f'{'R²':<10} {r2_1:>12.4f} {r2_2:>12.4f}')
print(f'{'RMSE':<10} {rmse_1:>12.4f} {rmse_2:>12.4f}')
print(f'{'MAE':<10} {mae_1:>12.4f} {mae_2:>12.4f}')
```

## 5. Clustering K-Means — Perfiles de estudiantes

```python
from pyspark.ml.clustering import KMeans
from pyspark.ml.feature import StandardScaler

assembler_km = VectorAssembler(
    inputCols=['horas_estudio','promedio_previo','horas_sueno','asistencia','uso_redes'],
    outputCol='features_raw',
    handleInvalid='skip'
)

scaler = StandardScaler(inputCol='features_raw', outputCol='features_scaled')

kmeans = KMeans(featuresCol='features_scaled', k=3, seed=42)

pipeline_km = Pipeline(stages=[assembler_km, scaler, kmeans])
model_km = pipeline_km.fit(df)
df_clusters = model_km.transform(df)

print('Distribución de estudiantes por perfil:')
df_clusters.groupBy('prediction').count().orderBy('prediction').show()

print('Promedios por perfil:')
df_clusters.groupBy('prediction').agg(
    F.round(F.avg('puntaje_final'), 2).alias('puntaje_promedio'),
    F.round(F.avg('horas_estudio'), 2).alias('horas_estudio_prom'),
    F.round(F.avg('promedio_previo'), 2).alias('promedio_previo_prom'),
    F.round(F.avg('asistencia'), 2).alias('asistencia_prom'),
    F.round(F.avg('uso_redes'), 2).alias('uso_redes_prom')
).orderBy('puntaje_promedio', ascending=False).show()
```

## 6. Guardar resultados en MinIO (S3)

```python
import boto3, json

s3 = boto3.client(
    's3',
    endpoint_url='http://minio:9000',
    aws_access_key_id='admin',
    aws_secret_access_key='password123'
)

# Guardar comparacion de modelos
resultados = {
    'modelo_1': {'r2': round(r2_1,4), 'rmse': round(rmse_1,4), 'mae': round(mae_1,4)},
    'modelo_2': {'r2': round(r2_2,4), 'rmse': round(rmse_2,4), 'mae': round(mae_2,4)}
}
s3.put_object(Bucket='academicaview', Key='resultados/comparacion_modelos.json',
    Body=json.dumps(resultados, indent=2))

# Guardar clusters como CSV
df_clusters.select('puntaje_final','horas_estudio','promedio_previo','asistencia','uso_redes','prediction') \
    .write.mode('overwrite') \
    .csv('s3a://academicaview/resultados/clusters', header=True)

print('Resultados guardados en MinIO:')
for obj in s3.list_objects(Bucket='academicaview', Prefix='resultados/')['Contents']:
    print(f"  {obj['Key']}")

spark.stop()
print('Análisis completado.')
```
